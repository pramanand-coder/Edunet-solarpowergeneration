```python
#plot distribution of power
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.histplot(data["generated_power_kw"],bins=30,kde=True)
plt.title("Distribution of Generated Power (kw)")
plt.xlabel("Generated Power (kw)")
plt.ylabel("Frequency")
plt.show()
```

```python
data.hist(bins=30)
```

```python
data[data.columns[:9]].hist(bins=30,figsize=(15,10))
plt.show()
```

```python
data[data.columns[9:]].hist(bins=30,figsize=(15,10))
plt.show()
```

```python
#Bivariate analysis
# scatter plot with target feature
plt.figure(figsize=(15,30))
for i,col in enumerate(data.columns):
    plt.subplot(7,3,i+1)
    plt.scatter(data[col],data['generated_power_kw'])
    plt.title(f'{col} Vs Generated Power Kw')
    plt.xlabel(col)
    plt.ylabel("Generated Power Kw")
plt.tight_layout()
plt.show()
```
